#### 利用 OpenCV API 进行图像融合


```python
import cv2
```

```python
img1 = cv2.imread("./photos/ori_left.jpg")
img2 = cv2.imread("./photos/ori_right.jpg")

stitcher = cv2.Stitcher_create()
status, result = stitcher.stitch([img1, img2])
cv2.imshow("stitched", result)
cv2.imwrite("opencv_airforce_stitched_result.jpg", result)
```
